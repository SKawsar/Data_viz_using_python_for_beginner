### What is customer churn?
Source: DataCamp and Mark Peterson

When an existing customer, user, player, subscriber or any kind of return client stops doing business or ends the relationship with a company is called customer churn.

**Contractual churn:** When a customer is under contract for a service and decides to cancel their service. Example: Cable TV, SaaS products (Software as a Service e.g. Dropbox).

**Voluntary churn:** When a user voluntarily cancels a service and includes prepaid cell phones, streaming subscriptions.

**Non-contractual churn:** When a customer is not under contract for a service and includes customer loyality at a retail location or online browsing. 

**Involuntary churn:** When a churn occurs not at the request of the customer. For example: credit card expiration, utilities being shut off by the provider.

Most likely, you as a customer have cancelled a service for a variety of reasons including lack of usage, poor service or better price. 

```python
# required library for data analysis
import pandas as pd
import numpy as np

# required librairies for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset: cellular usage dataset that consists of records of actual cell phone that include specific features such as

1. **Account_Length**: the number of days the customer has the subscription with the telecom company

2. **Vmail_Message**: the total number of voicemails the customer has sent

3. **Total_mins**: the total number of minutes the customer has talked over the phone

4. **CustServ_Calls**: the number of customer service calls the customer made

5. **Churn**: yes and no - indicating whether or not the customer has churned

6. **Intl_Plan**: yes and no - indicating whether or not the customer has international plan or not

7. **Vmail_Plan**: yes and no - indicating whether or not the customer has voicemail plan or not

8. **Total_calls**: the total number of calls the customer has made

9. **Total_charges**: the total amount of bill in $ the customer has paid

```python
# import the data 
# reading the csv file

# create a variable to save the input file
input_file = 'telecom_data.csv'

# load the input file
telecom = pd.read_csv(input_file)

# let's look inside the data
display(telecom.head())
```

### 4. How many customer's have international plan and churned?
### 5. How many customer's have voicemail plan and churned?

```python
# find the number of unique values in the 'Churn' and 'Intl_Plan' columns altogether
columns = ['Intl_Plan', 'Churn']
unique_Intl_Plan_churn = telecom[columns].value_counts()
print(unique__Intl_Plan_churn)

# let's visualize the number of unique values 
# in the 'Churn' and 'Intl_Plan' columns altogether in a 'pie' chart
unique_Intl_Plan_churn.plot(kind='pie')

# set the title of the image
plt.title("Customer Churn with or without international plan")
plt.ylabel("")

# display the figure
plt.show()
```

# Bar plot using pandas

```python
# find the number of unique values in the 'Churn' and 'Intl_Plan' columns altogether
columns_2 = ['Vmail_Plan', 'Churn']
unique_Vmail_Plan_churn = telecom[columns_2].value_counts()
print(unique__Vmail_Plan_churn)

# let's visualize the number of unique values 
# in the 'Churn' and 'Vmail_Plan' columns altogether in a 'bar' chart
unique_Vmail_Plan_churn.plot(kind='bar')

# set the title of the image
plt.title("Customer Churn with or without Voicemail plan")

# display the figure
plt.show()
```

# Bar plot using pandas

```python
plt.figure(figsize=(20,6))
plt.subplot(1, 3, 1)
unique_churn = telecom['Churn'].value_counts()
unique_churn.plot(kind='bar', color='red', alpha=0.8)
plt.ylabel("Count")
plt.title("Customer Churn")

plt.subplot(1, 3, 2)
unique_Intl_Plan = telecom['Intl_Plan'].value_counts()
unique_Intl_Plan.plot(kind='bar', color='red', alpha=0.5)
plt.ylabel("Count")
plt.title("International Plan")

plt.subplot(1, 3, 3)
unique_Vmail_Plan = telecom['Vmail_Plan'].value_counts()
unique_Vmail_Plan.plot(kind='bar', color='red', alpha=0.3)
plt.ylabel("Count")
plt.title("Voicemail Plan")
plt.tight_layout()
plt.show()
```

# Bar plot using Seaborn

```python
# barplot 
plt.figure(figsize=(20,6))

plt.subplot(1, 3, 1)
sns.histplot(data=telecom, x='Churn', color='r', alpha=0.5)
plt.title("Customer Churn")

plt.subplot(1, 3, 2)
sns.histplot(telecom['Intl_Plan'], color='g', alpha=0.5)
plt.title("International Plan")

plt.subplot(1, 3, 3)
sns.histplot(telecom['Vmail_Plan'], color='orange', alpha=0.5)
plt.title("Voicemail Plan")

plt.tight_layout()
plt.show()
```

# Histogram using Seaborn

```python
plt.figure(figsize=(20,15))

plt.subplot(3, 3, 1)
sns.histplot(data = telecom, x = 'Total_calls', binwidth=50)
plt.title("row = 1, column = 1, position = 1")

plt.subplot(3, 3, 2)
sns.histplot(data = telecom, x = 'Total_calls', binwidth=20, fill=False, color='green')
plt.title("row = 1, column = 2, position = 2")

plt.subplot(3, 3, 3)
sns.histplot(data = telecom, x = 'Total_calls', binwidth=20, stat='probability', color='green', alpha=0.5)
plt.title("row = 1, column = 3, position = 3")

plt.subplot(3, 3, 4)
sns.histplot(data = telecom, x = 'Total_calls', binwidth=20, stat='probability', cumulative=True)
plt.title("row = 2, column = 1, position = 4")

plt.subplot(3, 3, 5)
sns.histplot(data = telecom, x = 'Total_calls', binwidth=1, stat='probability', cumulative=True, hue='Churn')
plt.title("row = 2, column = 2, position = 5")

plt.subplot(3, 3, 6)
sns.histplot(data = telecom, x = 'Total_calls', binwidth=20, stat='probability', element='step')
plt.title("row = 2, column = 3, position = 6")

plt.subplot(3, 3, 7)
sns.histplot(data = telecom, x = 'Total_calls', binwidth=20, stat='probability', element='poly')
plt.title("row = 3, column = 1, position = 7")

plt.subplot(3, 3, 8)
sns.histplot(data = telecom, x = 'Total_calls', binwidth=20, stat='probability', kde=True)
plt.title("row = 3, column = 2, position = 8")

plt.subplot(3, 3, 9)
sns.histplot(data = telecom, x = 'Total_calls', binwidth=20, stat='probability', kde=True, hue='Churn')
plt.title("row = 3, column = 3, position = 9")

plt.show()
```
